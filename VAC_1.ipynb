```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
```

# **1. Change the number of shots to 100**

```python
# Create a quantum circuit with 1 qubit and 1 classical bit
qc = QuantumCircuit(1, 1)

qc.h(0)

qc.measure(0, 0)

simulator = Aer.get_backend('qasm_simulator')
result = simulator.run(qc, shots=100).result()
counts = result.get_counts(qc)

print("Task 1 - Results with 100 shots:", counts)
```

# **2. Add a second qubit, apply Hadamard gates to both qubits**

```python
# Create a quantum circuit with 2 qubits and 2 classical bits
qc2 = QuantumCircuit(2, 2)

qc2.h(0)
qc2.h(1)

qc2.measure([0,1], [0,1])


result2 = simulator.run(qc2, shots=100).result()
counts2 = result2.get_counts(qc2)

print("Task 2 - Results with 2 qubits in superposition:", counts2)
```

# **3. Replace the Hadamard gate with an X gate**

```python
# Create a quantum circuit with 1 qubit and 1 classical bit
qc3 = QuantumCircuit(1, 1)

qc3.x(0)
qc3.measure(0, 0)

result3 = simulator.run(qc3, shots=100).result()
counts3 = result3.get_counts(qc3)

print("Task 3 - Results with X gate:", counts3)
```
